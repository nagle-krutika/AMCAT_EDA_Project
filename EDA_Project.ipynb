```python
from datetime import datetime
```

```python
# Converting the DOB column to datetime.

df['DOB'] = pd.to_datetime(df['DOB'])
df['DOB']
```

```python
# Making a funtion which calculate the age from the given dob.

def calculate_age(birth_date):
    today = datetime.today()
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return age
```

```python
# making a new column 'age' to store the age of individuals.

df['age'] = df['DOB'].apply(calculate_age)
df.head(3)
```

```python
# Outlier treatment in salary column.

IQ1=df['Salary'].quantile(0.25)
IQ3=df['Salary'].quantile(0.75)

IQR=IQ3-IQ1
IQR

lower_limit=IQ1-1.5*IQR
upper_limit=IQ3+1.5*IQR

df['Salary'] = np.where((df['Salary'])> upper_limit,upper_limit,np.where((df['Salary']) < lower_limit, lower_limit,df['Salary']))
```

```python
# Top 10 cities with the highest average salaries.

avg_sal = df.groupby('JobCity')['Salary'].mean()
avg_salary = avg_sal.sort_values(ascending=False)

top_10_cities = avg_salary.head(20)

plt.figure(figsize=(10, 6))
top_10_cities.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities with Highest Average Salaries')
plt.xlabel('City')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Research Questions
-  Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.


```python
comp_df = df[df['Specialization'].str.contains('computer')]
comp_df['Specialization'].unique()
```

```python
new_df = comp_df[comp_df['Designation'].isin(['associate engineer','software engineer', 'hardware engineer','programmer analyst'])]
new_df
```

```python
new_df['Salary'].describe()
```

### Observations:

-  The mean salary of fresh graduates in Computer Science Engineering is approximately 342,965 lakhs, which is higher than the upper limit mentioned in the article (2.5-3 lakhs).

-  The minimum salary observed in the data is 80,000 lakhs, which is significantly lower than the lower limit mentioned in the article.

-  The maximum salary observed is 655,000 lakhs, which is substantially higher than the upper limit mentioned in the article.

-  The median salary (50th percentile) is 330,000 lakhs, which is higher than the upper limit mentioned in the article.

### Conclusion:

Based on the observatios, it appears that the earning potential for fresh graduates in Computer Science Engineering exceeds the range mentioned in the Times of India article. The salaries observed in the dataset are higher on average, with a significant portion of graduates earning above the upper limit stated in the article. Therefore, it can be concluded that the claim made in the article may not accurately reflect the current earning potential of fresh graduates in this field.
